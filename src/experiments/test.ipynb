```python
%load_ext autoreload
%autoreload 2

import torch
from lorenz import LorenzData
from norm_flow import AEVB
```

```python
def train_lorenz():
    # Setup data
    gen = LorenzData()
    T = 60
    num_trials = 100
    x_data, latents = gen.generate(num_trials, T)

    train_x = x_data[:80, :40, :] # Train on 80% trials, first 40 steps [cite: 125, 126]
    val_x = x_data[80:, :, :]     # Validate on 20% trials [cite: 125]

    mean = torch.zeros(3)
    var = torch.eye(3)


    model = AEVB(latent_dim=3, obs_dim=10, hidden_dim=64, base_mean=mean, base_var=var, batch_size=16, T=60)
    optimizer = torch.optim.Adam(model.parameters(), lr=1e-3)

    for epoch in range(1000):
        optimizer.zero_grad()
        loss = -model.elbo(train_x)
        loss.backward()
        optimizer.step()
        
        if epoch % 10 == 0:
            print(f"Epoch {epoch}, Loss: {loss.item()}")
```

```python
train_lorenz()
```
